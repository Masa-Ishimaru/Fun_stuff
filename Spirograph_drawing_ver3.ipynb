## Spirograph with animation
Reference: <br>
http://www.mathematische-basteleien.de/spirographs.htm

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
from IPython.display import clear_output, Image

%matplotlib inline
```

```python
# variables
x = np.arange(0, 36000, 0.1)
theta = x * (np.pi / 180.0)
```

#### Parameters to draw
<img src="spirograph_parameters.jpg" width="200px">

```python
# Examples
R = 9.0
r = 5.0
a = 12.0
```

```python
# Draw one picture of "Hypocycloids"
plt.figure(figsize = (5, 5))
spg_xi = (R - r) * np.cos((r / R) * theta) + a * np.cos((1 - (r / R)) * theta)
spg_yi = (R - r) * np.sin((r / R) * theta) - a * np.sin((1 - (r / R)) * theta)

plt.scatter(spg_xi, spg_yi, marker = '.', s = 0.01, alpha = 0.5)
plt.axis('off')
plt.show()
```

```python
# Draw one picture of "Epicycloids"
plt.figure(figsize = (5, 5))
spg_xi = (R + r) * np.cos((r / R) * theta) - a * np.cos((1 + (r / R)) * theta)
spg_yi = (R + r) * np.sin((r / R) * theta) - a * np.sin((1 + (r / R)) * theta)

plt.scatter(spg_xi, spg_yi, marker = '.', s = 0.01, alpha = 0.5)
plt.axis('off')
plt.show()
```

```python
# Draw continuously changing parameters R, r, and a
for a in range(3, 5, 1):
    for r in range(6, 10, 2):
        for R in range(15, 20):
            
                clear_output(True)
                plt.figure(figsize = (5, 5))

                spg_xi = (R - r) * np.cos((r / R) * theta) + a * np.cos((1 - (r / R)) * theta)
                spg_yi = (R - r) * np.sin((r / R) * theta) - a * np.sin((1 - (r / R)) * theta)

                plt.scatter(spg_xi, spg_yi, marker = '.', s = 0.01, alpha = 0.5)
                plt.axis('off')
                plt.show()
                print('          R =', R, ' r =', r, ' a =', a)
```

### Animation

```python
# Create GIF and mp4 animation and save them
ims = []
fig = plt.figure(figsize = (5, 5))

for R in range(30, 40, 2):
    for r in range(10, 30, 2):
        for a in range(5, 25, 5):
            
                plt.axis('off')
                spg_xi = (R - r) * np.cos((r / R) * theta) + a * np.cos((1 - (r / R)) * theta)
                spg_yi = (R - r) * np.sin((r / R) * theta) - a * np.sin((1 - (r / R)) * theta)
                
                im = plt.plot(spg_xi, spg_yi, linewidth = 0.5, alpha = 0.5)
                ims.append(im)
                
anim = animation.ArtistAnimation(fig, ims, interval = 100)
anim.save('spirograph_animation.gif', writer = 'pillow')
anim.save('spirograph_animation.mp4', writer = 'ffmpeg')
```
